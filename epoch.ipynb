```python
import pandas as pd
import matplotlib.pyplot as plt

epoch = pd.read_csv("data/epochvalid.csv")
epoch_transposed = epoch.drop(["Epochs"], axis=1).transpose()

colors = ["#7DB0A8","#9FD1F5","#FEE68E", "#D58A55"]
legend_names = ['Epoch 40', 'Epoch 60', 'Epoch 80', 'Epoch 100']

# Plotting
ax = epoch_transposed.plot(kind='bar', color=colors, figsize=(12, 6))

# Customizing the plot
plt.ylabel('ROC-AUC (%)')
plt.grid(True, linewidth=0.5, linestyle='--')
plt.xticks(rotation=0)

# Adding a legend
plt.legend(legend_names, bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=len(legend_names))

# # Adding values to each bar
# for container in ax.containers:
#     ax.bar_label(container)

plt.savefig("fig/epochvalid.png", dpi=600, bbox_inches='tight', transparent=False)
plt.show()
```

```python
import pandas as pd
epoch = pd.read_csv("data/epoch.csv")

# Setting the index to be the epochs column
epoch.set_index('Epochs', inplace=True)

# Plotting
epoch['Avg'].plot(kind='line', figsize=(6, 6))

# Customizing the plot
# plt.title('Line plot of Avg against Datasets')
plt.ylabel('ROC-AUC (%)')
plt.xlabel('Epochs')

# Adding a grid with custom size and weight
plt.grid(True, linewidth=0.5, linestyle='--')

# Rotating x-axis labels
plt.xticks(rotation=0)
plt.savefig("fig/epoch_avg.png",dpi=600)
# Displaying the plot
plt.show()
```
