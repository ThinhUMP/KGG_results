```python
import pandas as pd
epoch = pd.read_csv("data/epoch.csv")
epoch.head()
```

```python
import matplotlib.pyplot as plt# Transpose the DataFrame
epoch_transposed = epoch.transpose()

# Define a list of colors
colors = ["#008AD8", "#888B8D", "#EBBC4E", "#F2827F"]

# Plotting
epoch_transposed.plot(kind='bar', color=colors, figsize=(12, 6))

# Customizing the plot
plt.title('Bar plot of Epochs against Datasets')
plt.ylabel('ROC-AUC')
plt.xlabel('Datasets')
plt.grid(True, linewidth=0.2, linestyle='--', color='gray')
# Rotating x-axis labels
plt.xticks(rotation=0)

# Displaying the plot
plt.show()
```

```python
# Plotting
epoch_transposed['Avg'].plot(kind='line')

# Customizing the plot
plt.title('Line plot of Avg against Datasets')
plt.ylabel('Avg')
plt.xlabel('Datasets')

# Adding a grid with custom size and weight
plt.grid(True, linewidth=0.5, linestyle='--')

# Rotating x-axis labels
plt.xticks(rotation=0)

# Displaying the plot
plt.show()
```
