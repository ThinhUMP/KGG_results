```python
import pandas as pd
df = pd.read_excel("data/regression.xlsx")
df['RMSE'] = df[["ESOL", "FreeSolv", "Lipo"]].mean(axis=1)
df["MAE"] = df [["QM7", "QM8", "QM9"]].mean(axis=1)
df
```

```python
import matplotlib.pyplot as plt
# Sorting the DataFrame by the 'ESOL' column
for column in ["RMSE", "MAE"]:
    df = df.sort_values(column)

    # Creating a color array
    colors = ['#9FD1F5' if model != 'KGG' else '#D58A55' for model in df.Model]

    # Plotting with custom color
    plt.bar(df.Model, df[column], color=colors)

    # Customizing the plot
    plt.ylabel(f'{column}')

    # Adding a grid with custom size and weight
    plt.grid(True, linewidth=0.5, linestyle='--')

    # Rotating x-axis labels
    plt.xticks(rotation=70)
    plt.savefig(f'fig/{column}.png', dpi=600, bbox_inches='tight')
    # Displaying the plot
    plt.show()
```
