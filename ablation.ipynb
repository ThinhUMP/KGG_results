```python
import pandas as pd
ablation = pd.read_csv("data/ablation.csv")
import matplotlib.pyplot as plt# Transpose the DataFrame
ablation_transposed = ablation.drop(["Model", "Avg"], axis=1).transpose()

# Define a list of colors
colors = ["#7DB0A8","#9FD1F5","#FEE68E", "#D58A55", "#FFC1CC", "#64B372"]
legend_names = ['KGG-woPre', 'KGG', 'KGG-woBondType', 'KGG-woHyBo', 'KGG-HyBowoPre', 'KGG-woHybri']
# Plotting
ablation_transposed.plot(kind='bar', color=colors, figsize=(6, 6))

# Customizing the plot
# plt.title('Bar plot of Epochs against Datasets')
plt.ylabel('ROC-AUC (%)')
# plt.xlabel('Datasets')
plt.grid(True, linewidth=0.5, linestyle='--')
# Rotating x-axis labels
plt.xticks(rotation=0)
# Adding a legend
plt.legend(legend_names)
plt.savefig("fig/ablation.png", dpi=600)
# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
ablation = pd.read_csv("data/ablation.csv")
# Sorting the DataFrame by the 'Avg' column
ablation = ablation.sort_values('Avg')

# Creating a color array
colors = ['#D58A55' if model == 'KGG' else '#9FD1F5' for model in ablation.Model]

# Plotting with custom color
plt.bar(ablation.Model, ablation['Avg'], color=colors)

# Customizing the plot
plt.ylabel('ROC-AUC (%)')

# Adding a grid with custom size and weight
plt.grid(True, linewidth=0.5, linestyle='--')

# Rotating x-axis labels
plt.xticks(rotation=70)

# Saving the figure
plt.savefig("fig/ablation_avg.png", dpi=600)

# Displaying the plot
plt.show()
```
