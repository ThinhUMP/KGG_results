```python
import matplotlib.pyplot as plt
import numpy as np

# Data
metrics = ['BACE', 'BBBP', 'Tox21', 'ToxCast', 'SIDER', 'ClinTox']
ap_means = [85.0, 74.3, 34.3, 34.8, 63.9, 65.6]
ap_stds = [0.3, 0.8, 0.8, 0.9, 1.0, 1.3]
mcc_means = [48.7, 31.9, 27.3, 8.8, 12.7, 31.6]
mcc_stds = [1.2, 0.5, 1.0, 0.4, 0.7, 1.1]

x = np.arange(len(metrics))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 8))
rects1 = ax.bar(x, ap_means, width, label='AP', yerr=ap_stds, capsize=5, color='#9FD1F5')
# rects2 = ax.bar(x + width/2, mcc_means, width, label='MCC', yerr=mcc_stds, capsize=5, color='#D58A55')

# Increase the size of the font
ax.set_ylabel('%', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(metrics, fontsize=14)
ax.tick_params(axis='y', labelsize=14)

# Move the legend to the bottom
# ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=2, fontsize=14)

# Attach a text label above each bar in rects, displaying its height.
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.1f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 6),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontsize=14)

autolabel(rects1)
# autolabel(rects2)
plt.grid(True, linewidth=0.5, linestyle='--')
fig.tight_layout()
plt.savefig('fig/AP.png', dpi=600, transparent=False, bbox_inches='tight')
plt.show()
```
